Removing the unecessery

```python
data_ratings.drop("timestamp",axis=1,inplace=True)
data_ratings.head()
```

Merging dataset

```python
data_df=pd.merge(data_ratings,data_title,on="item_id")
data_df.head()
```

```python
data_df = data_df.pivot_table(index='user_id',columns='title',values='rating')
data_df
```

```python
data_df.fillna(0,inplace=True)
data_df
```

```python
movie_list=[]
for i in range(0,944):
    new_list=[]
    for j in range(0,1664):
        if data_df.values[i,j]!=0:
            movie=str(data_df.columns[j])
            new_list.append(movie)
    movie_list.append(new_list)
```

```python
len(movie_list)
```

```python
for i in range(len(movie_list)):
   print(movie_list[i])
```

```python
rules = apriori(transactions=movie_list, min_support=0.05, min_confidence=0.2, min_lift=6, max_length=2)
```

```python
print(rules)
```

```python
results = list(rules)
```

```python
print(len(results))
```

```python
def data_extractor(data):
    # Extract movies watched
    movies_watched = [i[2][0][0] for i in data]
    movie_df = pd.DataFrame(movies_watched, columns=['Movie_Watched'])

    # Extract also-watched items
    also_watched = [i[2][0][1] for i in data]
    also_watched_df = pd.DataFrame(also_watched, columns=['Also_Watched'])

    # Create a table for support values
    support_values = [i[1] for i in data]
    support_df = pd.DataFrame(support_values, columns=['Support'])

    # Create a table for confidence values
    confidence_values = [i[2][0][2] for i in data]
    confidence_df = pd.DataFrame(confidence_values, columns=['Confidence'])

    # Create a table for lift values
    lift_values = [i[2][0][3] for i in data]
    lift_df = pd.DataFrame(lift_values, columns=['Lift'])

    # Combine all tables into one final table
    result = pd.concat([movie_df, also_watched_df, support_df, confidence_df, lift_df], axis=1)
    return result

result_df = data_extractor(results)
```
